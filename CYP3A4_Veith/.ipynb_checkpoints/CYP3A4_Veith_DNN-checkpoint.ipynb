```python
class GaussianNoiseLayer(Layer):
    def __init__(self, stddev, **kwargs):
        super(GaussianNoiseLayer, self).__init__(**kwargs)
        self.stddev = stddev
    
    def call(self, inputs, training=None):
        if training:
            return inputs + tf.random.normal(tf.shape(inputs), mean=0.0, stddev=self.stddev)
        return inputs

    def get_config(self):
        config = super(GaussianNoiseLayer, self).get_config()
        config.update({'stddev': self.stddev})
        return config
    
####################################################################################################################

def masked_loss_function(y_true, y_pred):
    loss = K.categorical_crossentropy(y_true, y_pred)
    mask = K.cast(K.not_equal(K.argmax(y_true, axis=-1), padding_value), K.floatx())
    masked_loss = loss * mask
    masked_loss = K.sum(masked_loss) / K.maximum(K.sum(mask), K.epsilon())
    return masked_loss

####################################################################################################################

def masked_accuracy(y_true, y_pred):
    mask = K.cast(K.not_equal(K.argmax(y_true, axis=-1), padding_value), K.floatx())
    accuracy = K.cast(K.equal(K.argmax(y_true, axis=-1), K.argmax(y_pred, axis=-1)), K.floatx())
    masked_accuracy = K.sum(accuracy * mask) / K.maximum(K.sum(mask), K.epsilon())
    return masked_accuracy

####################################################################################################################



with open('/mimer/NOBACKUP/groups/theo-storage/Jalil/My-iLP/ZINC_PubChem/models_MolAI/char_to_int.pkl', 'rb') as f:
    char_to_int = pickle.load(f)
with open('/mimer/NOBACKUP/groups/theo-storage/Jalil/My-iLP/ZINC_PubChem/models_MolAI/int_to_char.pkl', 'rb') as f:
    int_to_char = pickle.load(f)
    
char_to_int_len = len(char_to_int)
max_smi_len = 111    #  3D-large:123   3D:123    2D:125     dimorph:117    MolAI:104

smiles_to_latent_model = load_model('/mimer/NOBACKUP/groups/theo-storage/Jalil/My-iLP/ZINC_PubChem/models_MolAI/smi2lat_epoch_6.h5')
latent_to_states_model = load_model('/mimer/NOBACKUP/groups/theo-storage/Jalil/My-iLP/ZINC_PubChem/models_MolAI/lat2states_epoch_6.h5')
model = load_model('/mimer/NOBACKUP/groups/theo-storage/Jalil/My-iLP/ZINC_PubChem/models_MolAI/Model_Trained_epoch_6.h5', custom_objects={'GaussianNoiseLayer': GaussianNoiseLayer,
                                                                                     'masked_loss_function' : masked_loss_function,
                                                                                     'masked_accuracy' : masked_accuracy})
```

```python
def vectorize(smiles):
    one_hot = np.zeros((smiles.shape[0], max_smi_len, char_to_int_len), dtype=np.int8)
    for i, smile in tqdm(enumerate(smiles)):
        
        # Encode the start char
        one_hot[i, 0, char_to_int["!"]] = 1
        
        # Encode the rest of the chars
        for j, c in enumerate(smile):
            if c in char_to_int:
                one_hot[i, j+1, char_to_int[c]] = 1
            else:
                continue  # Skip if char not in char_to_int
                
        # Encode end char
        one_hot[i, len(smile)+1, char_to_int["$"]] = 1
        
        one_hot[i, len(smile)+2:, char_to_int["%"]] = 1
        
    # Return two, one for input and the other for output
    return one_hot[:, 1:, :], one_hot[:, :-1, :]
```

```python
#define sampling model

input_shape = (max_smi_len-1, char_to_int_len)   #X_test.shape[1:]
output_dim = char_to_int_len           #Y_test.shape[-1]
lstm_dim = 1024

inf_decoder_inputs = Input(batch_shape=(1, 1, input_shape[1]))

inf_decoder_lstm1 = LSTM(lstm_dim,
                    return_sequences=True,
                    unroll=False,
                    stateful=True
                   )
inf_decoder_lstm2 = LSTM(lstm_dim,
                    return_sequences=True,
                    unroll=False,
                    stateful=True
                   )
inf_decoder_lstm3 = LSTM(lstm_dim,
                    return_sequences=True,
                    unroll=False,
                    stateful=True
                   )

inf_decoder_outputs = inf_decoder_lstm1(inf_decoder_inputs)
inf_decoder_outputs = inf_decoder_lstm2(inf_decoder_outputs)
inf_decoder_outputs = inf_decoder_lstm3(inf_decoder_outputs)

inf_decoder_dense = Dense(output_dim, activation='softmax')
inf_decoder_outputs = inf_decoder_dense(inf_decoder_outputs)

sample_model = Model(inf_decoder_inputs, inf_decoder_outputs)

#transfer Weights
sample_model.layers[1].set_weights(model.layers[12].get_weights())
sample_model.layers[2].set_weights(model.layers[28].get_weights())
sample_model.layers[3].set_weights(model.layers[44].get_weights())
sample_model.layers[4].set_weights(model.layers[58].get_weights())
```

```python
def latent_to_smiles(latent):
    
    #decode states and set Reset the LSTM cells with them
    states = latent_to_states_model.predict(latent)
    sample_model.layers[1].reset_states(states=[states[0], states[1]])
    sample_model.layers[2].reset_states(states=[states[2], states[3]])
    sample_model.layers[3].reset_states(states=[states[4], states[5]])
    
    #Prepare the input char
    startidx = char_to_int["!"]
    samplevec = np.zeros((1, 1, char_to_int_len))
    samplevec[0, 0, startidx] = 1
    smiles = ""
    
    #Loop and predict next char
    for i in range(max_smi_len-1):
        o = sample_model.predict(samplevec)
        sampleidx = np.argmax(o)
        samplechar = int_to_char[sampleidx]
        if samplechar != "$":
            smiles = smiles + int_to_char[sampleidx]
            samplevec = np.zeros((1, 1, char_to_int_len))
            samplevec[0, 0, sampleidx] = 1
        else:
            break
    return smiles#.replace("X", "Cl").replace("Y", "Br")
```

```python
# ANSI color codes
RESET = '\033[0m'
BLACK = '\033[30m'
RED = '\033[31m'
GREEN = '\033[32m'
YELLOW = '\033[33m'
BLUE = '\033[34m'
MAGENTA = '\033[35m'
CYAN = '\033[36m'
WHITE = '\033[37m'
BRIGHT_BLACK = '\033[90m'
BRIGHT_RED = '\033[91m'
BRIGHT_GREEN = '\033[92m'
BRIGHT_YELLOW = '\033[93m'
BRIGHT_BLUE = '\033[94m'
BRIGHT_MAGENTA = '\033[95m'
BRIGHT_CYAN = '\033[96m'
BRIGHT_WHITE = '\033[97m'

error = 0

for n in range(20):
    pred = latent_to_smiles(latent[n:n+1])
    if pred == filtered_rows['prepped_SMILES'][n]:
        print(f"{GREEN}{filtered_rows['prepped_SMILES'][n], pred}{RESET}")
    else:
        print(f"{RED}{filtered_rows['prepped_SMILES'][n], pred}{RESET}")
        error = error + 1

    print(100*'-')
print(error/len(range(20))*100)
```

```python
batch_size = 10000
error = 0
row = []

for batch_start in tqdm(range(0, len(X_test), batch_size)):
    batch_end = min(batch_start + batch_size, len(X_test))
    batch_indices = range(batch_start, batch_end)

    X_test_batch = X_test[batch_indices]
    Y_test_batch = Y_test[batch_indices]
    
    # Predict in batches
    batch_predictions = model.predict([X_test_batch, Y_test_batch])

    # Process each sample in the batch
    for i, idx in enumerate(batch_indices):
        v = batch_predictions[0][i]
        
        idxs2 = np.argmax(Y_test[idx:idx+1], axis=2)
        true = "".join([int_to_char[k] for k in idxs2[0]])[1:]
        true = true.replace("X", "Cl").replace("Y", "Br").replace("$", "").replace("%", "")
        
        idxs = np.argmax([v], axis=2)
        pred = "".join([int_to_char[h] for h in idxs[0]])[:-1]
        pred = pred.replace("X", "Cl").replace("Y", "Br").replace("$", "").replace("%", "")
        pred = pred[:len(true)]
        
        if true != pred:
            #print(f"\033[91m{true, pred}\033[0m")  # RED
            row.append([true, pred])
            error += 1

print((error / len(X_test)) * 100)
```

```python
x_train = train.drop("Target",axis=1)
y_train = train.pop('Target')  

x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
```

```python
def scheduler(epoch, lr):
    return lr * 0.9  # Reduce current learning rate by 5%
lr_scheduler = LearningRateScheduler(scheduler)
```

```python
bin_acc = [] 
bin_prec = []
bin_recall = []
bin_f1 = []
bin_auc = []  
bin_kappa = []
bin_mcc = []

folds = 5     #define the number of kfolds

for i in range(1):
    i+=1
    kf = KFold (folds, shuffle=True, random_state=i)     #kfold set up
    fold = 0
    
    for train, test in kf.split(x_train, y_train):     #train
        fold+=1
        print(f"Fold #{i}_{fold}")

        
        # define and fit model
        DNN_model = tf.keras.Sequential()
        DNN_model.add(tf.keras.layers.Dense(2048, input_shape=(512,), activation='relu', kernel_initializer='HeNormal'))
        DNN_model.add(tf.keras.layers.Dense(1024, activation='relu',kernel_initializer='HeNormal'))
        DNN_model.add(tf.keras.layers.Dense(1024, activation='relu',kernel_initializer='HeNormal'))
        #DNN_model.add(tf.keras.layers.Dense(50, activation='relu',kernel_initializer='HeNormal'))
        DNN_model.add(tf.keras.layers.Dense(1, activation='sigmoid',kernel_initializer='HeNormal'))

        DNN_model.compile(optimizer=tf.optimizers.Adam(0.0001), loss='binary_crossentropy', metrics=['binary_accuracy'])
        early_stopping = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True, monitor='val_loss')
        
        history = DNN_model.fit(x_train[train], y_train[train], epochs=100, validation_data=(x_train[test], y_train[test]), 
                                callbacks=[early_stopping, lr_scheduler], verbose=1, batch_size = 16)
        
        pred_prob = DNN_model.predict(x_train[test])
        pred_prob = np.squeeze(np.array(pred_prob))
        
        pred = (DNN_model.predict(x_train[test])> 0.5).astype('int32')
        pred = np.squeeze(np.array(pred))
        # save model to file

        filename  = './models/model_' + str(i) + '_' + str(fold) + '.h5'
        DNN_model.save(filename )
        print('Saved: %s' % filename )
        
        pd.concat([pd.DataFrame(y_train[test]),pd.DataFrame(pred_prob)], axis=1, ignore_index=True).to_csv("%s.csv" %filename, index=None)


        # calculate metrics and statistics
        
        accuracy = accuracy_score(y_train[test], pred)
        precision = precision_score(y_train[test], pred)
        recall = recall_score(y_train[test], pred)
        f1 = f1_score(y_train[test], pred)
        roc_auc = roc_auc_score(y_train[test], pred_prob)
        kappa = cohen_kappa_score(y_train[test], pred)
        mcc = matthews_corrcoef(y_train[test], pred)

        confusion_mat = pd.DataFrame(confusion_matrix(y_train[test], pred))

        # Print metrics with three digits after the decimal point
        print("Accuracy: {:.3f}".format(accuracy))
        print("Precision: {:.3f}".format(precision))
        print("Recall: {:.3f}".format(recall))
        print("F1 Score: {:.3f}".format(f1))
        print("ROC AUC Score: {:.3f}".format(roc_auc))
        print("Kappa Score: {:.3f}".format(kappa))
        print("MCC Score: {:.3f}".format(mcc))
        print("Confusion Matrix:")
        print(confusion_mat)
        print(pd.DataFrame(classification_report(y_train[test], pred, output_dict=True)))
        

        bin_acc.append(accuracy)
        bin_prec.append(precision)
        bin_recall.append(recall)
        bin_f1.append(f1)
        bin_auc.append(roc_auc)
        bin_kappa.append(kappa)    
        bin_mcc.append(mcc)
        
        %matplotlib inline
        # Assuming you have predicted probabilities 'y_pred_probs' and true labels 'y_true'
        fpr, tpr, thresholds = roc_curve(y_train[test], pred_prob)
        roc_auc = auc(fpr, tpr)

        # Plotting the ROC curve
        plt.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], 'k--')  # Plotting the random classifier curve (diagonal)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc='lower right')
        plt.show()
        print('------------------------------------------------------------------------------------------------------------')

    
print("Average_bin_Accuracy = %.3f (+/- %.3f)" % (np.mean(bin_acc), np.std(bin_acc)))
print("Average_bin_Precision = %.3f (+/- %.3f)" % (np.mean(bin_prec), np.std(bin_prec)))
print("Average_bin_Recall = %.3f (+/- %.3f)" % (np.mean(bin_recall), np.std(bin_recall)))
print("Average_bin_F1 = %.3f (+/- %.3f)" % (np.mean(bin_f1), np.std(bin_f1)))
print("Average_bin_AUC = %.3f (+/- %.3f)" % (np.mean(bin_auc), np.std(bin_auc)))
print("Average_bin_KAPPA = %.3f (+/- %.3f)" % (np.mean(bin_kappa), np.std(bin_kappa)))
print("Average_bin_MCC = %.3f (+/- %.3f)" % (np.mean(bin_mcc), np.std(bin_mcc)))
```
